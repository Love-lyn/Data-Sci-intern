```python
house_df = pd.read_csv("Housing.csv")
```

```python
titanic_df = pd.read_csv('cleaned_titanic.csv')
```

## Probability Distribution

```python
sns.histplot(house_df['price'], kde=True)
plt.title('Price Dsitribution')
plt.show()
```

```python
sns.histplot(titanic_df['Age'], kde=True)
plt.title('Age Distribution')
plt.show()

sns.histplot(titanic_df['Pclass'], kde=True)
plt.title('Passenger Class Distribution')

plt.show()

```

## Hypothesis Testing

```python
# For the Titanic Dataset
male_survival = titanic_df[titanic_df['Sex'] == 'male']['Survived']
female_survival = titanic_df[titanic_df['Sex'] == 'female']['Survived']

ttest_ind(male_survival, female_survival)
```

```python
# For the House prices Dataset

furnished = house_df[house_df['furnishingstatus'] == 'furnished']['price']
unfurnished = house_df[house_df['furnishingstatus'] == 'unfurnished']['price']

ttest_ind(furnished, unfurnished)
```

## Correlation

```python
corr = house_df.corr(numeric_only = True)

sns.heatmap(corr, annot  =True, cmap = 'coolwarm')
plt.title('House Prices Correlation Matrix')
plt.show()
```

```python
corr = titanic_df.corr(numeric_only = True)

sns.heatmap(corr, annot  =True, cmap = 'coolwarm')
plt.title('Titanic Correlation Matrix')
plt.show()
```
